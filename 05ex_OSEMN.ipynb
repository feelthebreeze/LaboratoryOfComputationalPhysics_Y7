2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
rmatrix = np.random.random(size=(5,5))
rmatrix
```

```python
with open('data.txt', 'w') as f:
    for line in rmatrix:
        f.write("".join(str(line)) + "\n")
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
data_text = open('data.txt', 'r')
print(data_text.read())
```

```python
with open('data.txt', 'r') as infile, open('data.csv', 'w') as outfile:
        stripped = (line.strip() for line in infile)
        replaced1 = (strip.replace("[", "") for strip in stripped)
        replaced2 = (strip.replace("]", "") for strip in replaced1)
        lines = (line.split(",") for line in replaced2 if line)
        writer = csv.writer(outfile)
        writer.writerows(lines)
```
